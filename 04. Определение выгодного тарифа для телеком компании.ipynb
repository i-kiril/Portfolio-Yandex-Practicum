**Таблица users (информация о пользователях):**
- user_id — уникальный идентификатор пользователя
- first_name — имя пользователя
- last_name — фамилия пользователя

**Таблица messages (информация о сообщениях):**
- id — уникальный номер сообщения

- rub_per_minute — стоимость минуты разговора сверх тарифного пакета (например, если в тарифе 100 минут разговора в месяц, то со 101 минуты будет взиматься плата)
- rub_per_message — стоимость отправки сообщения сверх тарифного пакета
- rub_per_gb — стоимость дополнительного гигабайта интернет-трафика сверх тарифного пакета (1 гигабайт = 1024 мегабайта)

Создаем функцию расчета помесячной выручки по звонкам  

```python
def revenue_calls (df):
    tariff = df['tariff']
    if tariff == 'smart':
        extra_minutes = df['call_minutes'] - df_tariffs.loc[0,'minutes_included']
        if extra_minutes > 0 :
            call_cost =  df_tariffs.loc[0,'rub_monthly_fee'] + extra_minutes*df_tariffs.loc[0,'rub_per_minute']
        else :
            call_cost = df_tariffs.loc[0,'rub_monthly_fee']
    elif tariff == 'ultra':
        extra_minutes = df['call_minutes'] - df_tariffs.loc[1,'minutes_included']
        if extra_minutes > 0 :
            call_cost =  df_tariffs.loc[1,'rub_monthly_fee'] + extra_minutes*df_tariffs.loc[1,'rub_per_minute']
        else :
            call_cost = df_tariffs.loc[1,'rub_monthly_fee']
    return call_cost
```

Создаем функцию расчета помесячной выручки по сообщениям

```python
def revenue_sms (df):
    tariff = df['tariff']
    if tariff == 'smart':
        extra_sms = df['sms_count'] - df_tariffs.loc[0, 'messages_included']
        if extra_sms > 0 :
            cost_sms =  extra_sms*df_tariffs.loc[0, 'rub_per_message']
        else :
            cost_sms = 0
    elif tariff == 'ultra':
        extra_sms = df['sms_count'] - df_tariffs.loc[1, 'messages_included']
        if extra_sms > 0 :
            cost_sms =  extra_sms*df_tariffs.loc[1, 'rub_per_message']
        else :
            cost_sms = 0
    return cost_sms
```

Создаем функцию расчета помесячной выручки интернет трафика

```python
def revenue_gb (df):
    tariff = df['tariff']
    if tariff == 'smart':
        extra_gb = df['gb_used'] - df_tariffs.loc[0, 'mb_per_month_included'] / 1024
        if extra_gb > 0 :
            cost_gb =  (extra_gb)*df_tariffs.loc[0, 'rub_per_gb']
        else :
            cost_gb = 0
    elif tariff == 'ultra':
        extra_gb = df['gb_used'] - df_tariffs.loc[1, 'mb_per_month_included'] / 1024
        if extra_gb > 0 :
            cost_gb =  (extra_gb)*df_tariffs.loc[1, 'rub_per_gb']
        else :
            cost_gb = 0
    return cost_gb
```

Создаем столбец с суммой помесячной выручки пользователей по звонкам, сообщениям и интернет трафика

```python
full_data['revenue_month'] = ( full_data.apply(revenue_calls, axis = 1)+

 full_data.apply(revenue_sms, axis = 1) +

 full_data.apply(revenue_gb, axis = 1) )

full_data.head()
```

Добавляем фрем данных с новым списком в качестве новой строки

```python
def add_row(df,ls):
    num_el = len(ls)
    new_row = pd.DataFrame(np.array(ls).reshape(1,num_el), columns = list(df.columns))
    df = df.append(new_row, ignore_index=True)
    return df
```
